```python
import matplotlib.pyplot as plt
from PIL import Image
fname = r'Afridi.jpg'
Image = Image.open(fname).convert('L')
plt.imshow(Image,cmap = 'gray')
plt.title("Liaquat Ali Afridi")
plt.xlabel("Length")
plt.ylabel("Breadth")
plt.grid()
plt.show()
```

# Matplotlib - pie

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['English','Hindi','Maths','Science']
plt.pie(x,labels = y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ['English','Hindi','Maths','Science']
c = ['yellow','magenta','brown','blue']
plt.pie(x,labels = y, colors = c)
plt.legend()
plt.show()
```
