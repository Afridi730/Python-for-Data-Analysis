```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Matplotlib

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
plt.plot(var,var1)
plt.show()
```

# Seaborn Lib

```python
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
sns.lineplot(x = var,y = var1)
plt.show()
```

# Plotting Dictionary

```python
import pandas as pd
var = [1,2,3,4,5,6,7,8]
var1 = [8,7,6,5,4,3,2,1]
df = pd.DataFrame( { "var":var,"var1":var1})
df
```

```python
sns.lineplot(x= 'var', y='var1', data = df)
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
df3 = sns.load_dataset("planets")
df3
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1 )
plt.show
```

# To separate male and female categories

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex" )
plt.show
```

# Change of color

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, hue = "sex", palette ="rocket_r" )
plt.show
```

# Change of Style

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df1, style = "sex", palette ="rocket_r" )
plt.show
```

# Markers

```python
df = df1.head(20)
df
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df, style = "sex", palette ="rocket_r",markers = ["o", ">"] )
plt.show
```

# Grid & Title

```python

sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm", data = df, style = "sex", palette ="rocket_r",markers = ["o", ">"] )
plt.show

plt.grid()
plt.title("This is a Line Graph Related to Penguins")
```
