# Seaborn Library Part 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Get dataset from Github

```python
df1 = sns.load_dataset("penguins")
```

```python
df1["island"]
```

# Bar plot in Seaborn Library

```python
sns.barplot( x = df1.island , y = df1.bill_length_mm)
```

```python
sns.barplot( x = "island", y = "bill_length_mm", data = df1)
```

# Dividing Bars as per Category using "hue"  parameter

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df1 , hue = "sex")
```

# Order Parameter

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm",data = df1 , hue = "sex", order = order1)
plt.show()
    
```

# Changing of order of Bars using hue_order parameter

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex" , hue_order = ["Female", "Male"],order = order1)
```

# Changing of Colors

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1,order = order1, color = 'red')
```

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex" ,order = order1 , palette = 'Accent')
```

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex" , palette = "prism",order = order1)
```

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex" , palette = "gist_earth",order = order1)
```

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex" ,palette = "gist_heat",order = order1)
```

# Saturation Parameter

```python
order1 = ["Dream","Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex" ,saturation = .8 , palette = "prism",order = order1)
```

# Histogram in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df1
sns.displot(df1["flipper_length_mm"])
```

# Bins Parameter

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240])
```

# Color in Histogram

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240], color = 'green')
```

# Kernel density kde

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240], color = 'red' , kde = True)
```
