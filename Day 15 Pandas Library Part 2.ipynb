# Mathematical Operations

```python
m1 = pd.Series([100,200,300,400,500], index = [1,2,3,4,5])
print(m1)
```

```python
m2 = pd.Series([600,700,800,900,1000], index =[1,2,3,4,5])
print(m2)
```

```python
m3 = pd.concat([m1,m2])
print(m3)
```
