# Matplotlib Library Part 1

```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x = [1,2,3,4,]
y = [5,6,7,8]
plt.plot(x,y)
```

# Change color of graph

```python
x = [1,2,3,4,]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
```

# Random values graph

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y)
```

# Markers on the graph

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(x,y,marker ='o')
```

# Plotting x and y separately

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]
plt.plot(y,marker ='>')
plt.plot(x,marker ='o')
         
```

# Bar Graph

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
```

# Change color of bars

```python
x=[1,2,3,4]
y=[5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y,color = c )
```

# Horizontal Representation of Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array(['A', 'B','C','D'])
y = np.array ([3,8,1,10])
plt.barh(x,y)
```

```python
plt.barh(x,y,color = 'red')
```

```python
plt.barh(x,y,color ="#4CAF50")
```

```python
plt.bar(x,y,color='r')
```

# Changing Width of Bar

```python
plt.bar(x,y,color='red',width = .2)
```

# Changing Height of Bar

```python
plt.barh(x,y,color = 'green' , height = 0.1)
```

# Scatter Plot

# Title

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.scatter(x,y)
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.scatter(x,y)
plt.title("Scatter Plot")
```

# Label

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.scatter(x,y)

plt.xlabel('Months')
plt.ylabel('Numbers')
```

# Color

```python
plt.scatter(x,y,color='y')
```

# Separate Colors

```python
c = ['red','magenta','green','yellow','blue','brown']
plt.scatter(x,y,color = c)
```

# Size of dots in plot

```python
plt.scatter(x,y,color=c,s=150)
```
